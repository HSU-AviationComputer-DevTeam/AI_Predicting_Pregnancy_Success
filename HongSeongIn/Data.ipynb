```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

#   
plt.rcParams['font.family'] = 'NotoSansCJK-Regular'
plt.rcParams['axes.unicode_minus'] = False

#  
data = pd.read_csv('train.csv' encoding='utf-8-sig')

#    
numeric_columns = data.select_dtypes(include=['number'])
correlation_matrix = numeric_columns.corr()

#   
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            cbar=True,
            square=True)
plt.title('    ', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# _     
success_correlations = correlation_matrix['_'].sort_values(key=abs, ascending=False)
print("\
===    (  ) ===")
print(success_correlations)
```
