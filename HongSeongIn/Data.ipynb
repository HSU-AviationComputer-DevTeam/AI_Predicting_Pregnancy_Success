```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NotoSansCJK-Regular'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
data = pd.read_csv('train.csv' encoding='utf-8-sig')

# 수치형 컬럼만 상관관계 분석
numeric_columns = data.select_dtypes(include=['number'])
correlation_matrix = numeric_columns.corr()

# 상관관계 히트맵 그리기
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            cbar=True,
            square=True)
plt.title('임신 성공과 변수들 간의 상관관계', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 임신_성공과의 상관관계만 추출하여 절댓값 기준으로 정렬
success_correlations = correlation_matrix['임신_성공'].sort_values(key=abs, ascending=False)
print("\
=== 임신 성공과의 상관관계 (절댓값 기준 정렬) ===")
print(success_correlations)
```
